```r
if(!dir.exists("files")) dir.create("files")
url <- "https://thredds.daac.ornl.gov/thredds/fileServer/ornldaac/2130/"
options(timeout = 1200)
```

```r
prcp_files <- map(
  seq(1980, 2022, 1),
  \(x) paste0(url, "daymet_v4_prcp_annttl_na_", x, ".nc")
)
map(
  prcp_files,
  function(x) {
    download.file(
      url = x,
      destfile = paste0("files/", last(unlist(strsplit(x, "/"))))
    )
  }
)
# map(
#   prcp_files,
#   function(x) {
#     print(
#       paste0("files/", last(unlist(strsplit(x, "/"))))
#     )
#   }
# )
```
